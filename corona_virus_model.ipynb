# Corona Virus Model

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import sys

set = {}
set['population_size'] = 150
set['total_generation'] = 50
set['total_virus'] = 100
set['x_min'] = 0.01
set['x_max'] = 0.99
set['y_min'] = 0.01
set['y_max'] = 0.99

#Population Class
class Population:
  def __init__(self, set):
    self.x_pos = random.uniform(set['x_min'],set['x_max'])
    self.y_pos = random.uniform(set['x_min'],set['x_max'])
    self.speed = random.uniform(5,10)
    self.health = random.uniform(0,10)
    self.age = random.uniform(10,70)
    self.gender = random.uniform(0,1)
    self.immunity = random.uniform(0,10)

  def plot_population(self, set, ax):
    shape = Circle([self.x_pos, self.y_pos], 0.005, facecolor = 'green',zorder = 1)
    ax.add_artist(shape)
    pass

#Virus Class
class Virus:
  def __init__(self):
    self.transmitive_rate = random.uniform(0,1)
    self.mutation_rate = 0.10



def plot_set():
  fig, ax= plt.subplots()
  fig.set_size_inches(9.6, 9.6)
  for population in range(0, set['population_size']):
    population = Population(set)
    population.plot_population(set, ax)
    
  # MISC PLOT SETTINGS
  ax.set_aspect('equal')
  frame = plt.gca()
  frame.axes.get_xaxis().set_ticks([])
  frame.axes.get_yaxis().set_ticks([])
  plt.figtext(0.025, 0.95, r'GENERATION: ' + "50")
  plt.figtext(0.025, 0.90, r'T_STEP: ' + "000")
  plt.show()


plot_set()
```
